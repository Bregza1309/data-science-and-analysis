<h2>Extracting the nearest neighbors</h2>

Load Necessary modules

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
import pandas as pd
import seaborn as sb
```

Prepare Data

```python
X = np.array([[2.1, 1.3], [1.3, 3.2], [2.9, 2.5], [2.7, 5.4], [3.8, 0.9],
 [7.3, 2.1], [4.2, 6.5], [3.8, 3.7], [2.5, 4.1], [3.4, 1.9],
 [5.7, 3.5], [6.1, 4.3], [5.1, 2.2], [6.2, 1.1]])
X.shape
```

Define the number of nearest neighbors you want to extract:

```python
neighbors = 5
```

Define a test datapoint that will be used to extract the K nearest neighbors

```python
test_datapoint = np.array([[4.3, 2.7]])
```

Plot the input data using circular shaped black markers:


```python
fig,ax = plt.subplots(figsize = (10,8))
ax.set_title('Original Scatter of Data')
arg = {'x':X[:,0],'y':X[:,1],'marker':'o'}
sb.scatterplot(**arg)
```

Create and train a K Nearest Neighbors model using the input data. Use this model to
extract the nearest neighbors to our test data point:

```python
knn_model = NearestNeighbors(n_neighbors = neighbors,algorithm = 'ball_tree').fit(X)
distances,indices = knn_model.kneighbors(test_datapoint)
```

Print the nearest neighbors extracted from the model:Print the nearest neighbors extracted from the model:

```python
print('\nK Nearest Neighbors:')
for rank , index in enumerate(indices[0][:neighbors],start = 1):
    print(str(rank) + "==>",X[index])
```

Visualize the nearest neighbors:

```python
fig,axes = plt.subplots(1,2,figsize = (15,6))
axes[0].set_title('Original Scatter of Data')
arg = {'x':X[:,0],'y':X[:,1],'marker':'o','ax':axes[0]}
sb.scatterplot(**arg)
axes[1].set_title('Scatter of Data Showing the Nearest Neighbors')
arg = {'x':X[:,0],'y':X[:,1],'marker':'o','ax':axes[1]}
sb.scatterplot(**arg)

arg = {'x':X[indices][0][:][:,0],'y':X[indices][0][:][:,1],'marker':'o','ax':axes[1],'color':'#000000','s':300}
sb.scatterplot(**arg)

arg = {'x':test_datapoint[:,0],'y':test_datapoint[:,1],'marker':'o','ax':axes[1],'color':'#ff0000','s':200}
sb.scatterplot(**arg)
```
